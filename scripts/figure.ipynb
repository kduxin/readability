```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use(['science'])
```

```python
df = pd.read_csv('mixture_weight.csv')
del df['Name']
del df['_wandb']
df.columns = [col.replace('mixer_weight/', '') for col in df.columns]
df = df[sorted(df.columns)]
df
```

```python
# for i, col in enumerate(df):
fig, ax = plt.subplots(1, 1, figsize=(5,4), dpi=200)
sns.boxplot(data=df)

ax.set_ylabel('Weight', fontdict={'size':16})
ax.set_xlabel('Layer', fontdict={'size':16})
```
